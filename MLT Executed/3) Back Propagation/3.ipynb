```python

import numpy as np

X = np.array(([2,9],[1,5],[3,6]), dtype = float)
Y = np.array(([92],[86],[89]), dtype = float)
X = X/np.amax(X,axis=0) 
Y = Y/100

def sigmoid(x):
    return 1/(1 + np.exp(-x))
def derivatives(x):
    return x*(1-x)

epoch = 20
lr=0.1
ini=2
hn=3
on=1

wh = np.random.uniform(size=(ini,hn))
bh=np.random.uniform(size=(1,hn))

wout = np.random.uniform(size=(hn,on))
bout=np.random.uniform(size=(1,on))

for i in range(epoch):
    
    #o/p forward prop
    hinp = np.dot(X,wh) + bh
    hlayer = sigmoid(hinp)
    oinp = np.dot(hlayer,wout) + bout
    output = sigmoid(oinp)
 
    
    
    #error back prop.
    eo = (Y - output)*derivatives(output)
    eh = eo.dot(wout.T)*derivatives(hlayer)
    
    
    
    #LC addition to wts
    wh+=X.T.dot(eh)*lr
    wout+=hlayer.T.dot(eo)*lr
    print("epoch=",i,"lr=",lr,"error=",eh)
    
 
print("Input\n" ,X)
print("Actual Output\n", Y)
print("Predicted Output\n" ,output)
```
